```python
bucket_name="spark_dataset1"
path=f"gs://{bucket_name}/brewery_dataa/brewery_data.csv"
data=spark.read.csv(path, header=True)
data.show()
```

```python
#1 std,avg for a integer column
from pyspark.sql.functions import sum, avg,std
data.select(avg("Color").alias("avg_Color"),std("Color").alias("std_Color")).show()
```

```python
#2 Top 5 frequency in any one of the string based column
from pyspark.sql.functions import col
freq_data = data.groupBy("Location").count()
top5_freq = freq_data.orderBy(col("count").desc()).limit(5).show()
```
